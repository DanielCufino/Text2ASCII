```python
import numpy as np
import pandas as pd
import os
import PIL, PIL.Image
import base64
import io, math, json

import matplotlib as mpl
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
dataset = pd.read_csv("full_emoji_parsed.csv")
```

```python
dataset.head()
```

```python
types = list(dataset.columns[11 +5:11 +16])
types
```

```python
raw_data = []

for row in range(dataset.shape[0]):
    entry = dataset.iloc[row]
    for t in types:
        company = t.split("_")[0]
        description = entry['name']
        if isinstance(entry[t], str):

            json_entry = {
                "prompt": f'Please generate ASCII art of an emoji of {description}',
                "text": entry[t]
            }
            raw_data.append(json_entry)
```

```python
raw_data_string = json.dumps(raw_data)
raw_data_file = open("raw_data.json", "w")
raw_data_file.write(raw_data_string)
raw_data_file.close()
```

```python
with open('raw_data.json', 'r') as file:
    test = json.load(file)
    print(test[0])

p = test[0]['prompt']

t = test[0]['text']
print(f'<BOS> {p} \n\n <RES>: \n\n {t} \n\n <EOS>')
```
